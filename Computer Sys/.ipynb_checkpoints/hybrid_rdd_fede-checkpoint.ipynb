```python
from pyspark import SparkContext
import csv
import matplotlib.pyplot as pyplot

sc = SparkContext.getOrCreate()

train_rdd = sc.textFile("data/train.csv")
icm_rdd = sc.textFile("data/icm_fede.csv")
test_rdd= sc.textFile("data/target_users.csv")

train_header = train_rdd.first()
icm_header = icm_rdd.first()
test_header= test_rdd.first()

train_clean_data = train_rdd.filter(lambda x: x != train_header).map(lambda line: line.split(',')).map(lambda x: (int(x[0]), int(x[1]), float(x[2]))).cache()
icm_clean_data = icm_rdd.filter(lambda x: x != icm_header).map(lambda line: line.split(',')).map(lambda x: (int(x[0]), int(x[1]))).cache()
test_clean_data= test_rdd.filter(lambda x: x != test_header).map(lambda line: line.split(',')).cache()
```

```python
icm_clean_data.take(1)
```

```python
icm2 = icm_clean_data.groupByKey().map(lambda x: (x[0], list(x[1]))).sortByKey()
```

```python
items = icm2.map(lambda x: x[0]).collect()
```

```python
features = icm2.map(lambda x: len(x[1])).collect()
```

```python
pyplot.plot(items, features)
pyplot.ylabel('features')
pyplot.xlabel('items')
pyplot.show()
```

```python
train_items = train_clean_data.map(lambda x: (x[1], x[2])).groupByKey().map(lambda x: (x[0], list(x[1]))).sortByKey()
```

```python
items2 = train_items.map(lambda x: x[0]).collect()
```

```python
ratings = train_items.map(lambda x: len(x[1])).collect()
```

```python
pyplot.plot(items2, ratings)
pyplot.ylabel('ratings')
pyplot.xlabel('items')
pyplot.show()
```
